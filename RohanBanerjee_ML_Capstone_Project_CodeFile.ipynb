**Step 1: Load and Inspect Dataset**

```python
import pandas as pd
from pathlib import Path

csv_path = "/content/expected_ctc.csv"
df = pd.read_csv(csv_path)

print("Shape:", df.shape)
print("\nColumns:", list(df.columns))
print("\nTarget candidates:", [c for c in df.columns if "expected" in c.lower() or "ctc" in c.lower()])

print("\nMissing values (top 20):")
print(df.isnull().sum().sort_values(ascending=False).head(20))

print("\nNumber of duplicate rows:", int(df.duplicated().sum()))
display(df.head())
display(df.describe(include='all').T)
```

**Step 2: Handle Missing Values and Duplicates**

```python
import numpy as np

df.info()
# Drop ID columns
df = df.drop(columns=['IDX', 'Applicant_ID'])

# Drop duplicates if any
df = df.drop_duplicates()

# Separate numeric and categorical columns
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Impute numeric columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        skewness = df[col].skew()
        if abs(skewness) > 0.5:
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)

# Impute categorical columns with mode
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Final check
print("Missing values after imputation:")
print(df.isnull().sum().sum())  # should be 0
print("Shape after cleaning:", df.shape)
```

**Step 4: Iterative Outlier Removal**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Boxplots BEFORE removal ---
num_cols = df.select_dtypes(include=['number']).columns

plt.figure(figsize=(20, 3*(len(num_cols)//4+1)))
df[num_cols].plot(kind='box', subplots=True, layout=(len(num_cols)//4+1, 4),
                  figsize=(20, 3*(len(num_cols)//4+1)), patch_artist=True)
plt.suptitle("Boxplots BEFORE Outlier Removal", fontsize=18)
plt.show()


# --- 2. Iterative Outlier Removal ---
def iterative_outlier_removal(data, numeric_columns):
    df_clean = data.copy()
    iteration = 0
    while True:
        outlier_found = False
        for col in numeric_columns:
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            before_shape = df_clean.shape[0]
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
            after_shape = df_clean.shape[0]

            if before_shape != after_shape:
                outlier_found = True

        iteration += 1
        if not outlier_found:
            break
        print(f"Iteration {iteration}: Remaining rows = {df_clean.shape[0]}")
    return df_clean

df = iterative_outlier_removal(df, num_cols)
print("Shape after outlier removal:", df.shape)


# --- 3. Boxplots AFTER removal ---
plt.figure(figsize=(20, 3*(len(num_cols)//4+1)))
df[num_cols].plot(kind='box', subplots=True, layout=(len(num_cols)//4+1, 4),
                  figsize=(20, 3*(len(num_cols)//4+1)), patch_artist=True)
plt.suptitle("Boxplots AFTER Outlier Removal", fontsize=18)
plt.show()
```

**Step 5: One-Hot Encoding**

```python
import pandas as pd

# Separate features (X) and target (y)
target_col = "Expected_CTC"
X = df.drop(columns=[target_col])
y = df[target_col]

# One-Hot Encode categorical columns
X_encoded = pd.get_dummies(X, drop_first=True)  # drop_first=True to avoid dummy variable trap

print("Shape before encoding:", X.shape)
print("Shape after encoding:", X_encoded.shape)
```

**Step 6: Train-Test Split**
            +
**Step 7: Normalization & Scaling**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#  Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

# Normalization / Scaling
scaler = StandardScaler()

# Fit scaler only on training data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling done. Feature means after scaling (train):", X_train_scaled.mean(axis=0)[:5])
print("Feature stds after scaling (train):", X_train_scaled.std(axis=0)[:5])
```

**Step 8: Training and Evaluating Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Train Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)

# Predictions
y_train_pred = lin_reg.predict(X_train_scaled)
y_test_pred = lin_reg.predict(X_test_scaled)

# Metrics
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

print("Linear Regression Results")
print(f"Train RMSE: {train_rmse:.2f}")
print(f"Test RMSE: {test_rmse:.2f}")
print(f"Train R²: {train_r2:.4f}")
print(f"Test R²: {test_r2:.4f}")
```

**Step 9: Training and Evaluating Random Forest Regressor**                                
[Comparison between the RMSE and R-square for both models also shown]

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest
rf_reg = RandomForestRegressor(
    n_estimators=200,      # number of trees
    random_state=42,
    n_jobs=-1              # use all CPU cores
)
rf_reg.fit(X_train_scaled, y_train)

# Predictions
y_train_rf = rf_reg.predict(X_train_scaled)
y_test_rf = rf_reg.predict(X_test_scaled)

# Metrics
train_rmse_rf = np.sqrt(mean_squared_error(y_train, y_train_rf))
test_rmse_rf = np.sqrt(mean_squared_error(y_test, y_test_rf))

train_r2_rf = r2_score(y_train, y_train_rf)
test_r2_rf = r2_score(y_test, y_test_rf)

print(" Random Forest Regressor Results")
print(f"Train RMSE: {train_rmse_rf:.2f}")
print(f"Test RMSE: {test_rmse_rf:.2f}")
print(f"Train R²: {train_r2_rf:.4f}")
print(f"Test R²: {test_r2_rf:.4f}")

# Comparison
print("\n Performance Comparison:")
print(f"Linear Regression - Test R²: {test_r2:.4f} | RMSE: {test_rmse:.2f}")
print(f"Random Forest    - Test R²: {test_r2_rf:.4f} | RMSE: {test_rmse_rf:.2f}")
```

**Step 10: Visual Comparisons for both Models**
